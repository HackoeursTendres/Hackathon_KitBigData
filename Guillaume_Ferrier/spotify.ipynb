```python
# track input
#inputracks=pd.read_csv("songs.csv",names=["artist","title"], header=0)
#inputracks=pd.read_csv('BestHits.csv',header=0)
inputracks=pd.read_csv("youngTests.csv",names=["artist","title"], header=0)

# Reset output
dfadd = pd.DataFrame(columns=[
    
"artistsrc","titlesrc",
    
'danceability',
 'energy',
 'key',
 'loudness',
 'mode',
 'speechiness',
 'acousticness',
 'instrumentalness',
 'liveness',
 'valence',
 'tempo',
 'type',
 'id',
 'uri',
 'track_href',
 'analysis_url',
 'duration_ms',
 'time_signature'])






cptr = 0
for index, row in inputracks.iterrows():
    #print(row)
    tmpartist=row['artist']
    tmptitle=row['title']
#    tmpartist=row['Artiste']
#    tmptitle=row['Titre']

    tmpquery=tmpartist+" "+tmptitle

    r = requests.get(querytrack.format(urllib.parse.urlencode({'q':tmpquery}).replace("+","%20")  ),headers=headers)
    #print ( querytrack.format(urllib.parse.urlencode({'q':tmpquery}).replace("+","%20") ) )
    if r.status_code!=200:
        print('API search KO for '+str(r.status_code)+' '+tmpartist + ' '+tmptitle)
        continue

    outquery = json.loads(r.content.decode())['tracks']
    if outquery['total']==0:
        print('search Empty for '+tmpartist + ' '+tmptitle)
        continue
 
    songid=outquery['items'][0]['id']
    r3 = requests.get(audiofeat.format(songid),headers=headers)
    if r.status_code!=200:
        print('API song KO for '+tmpartist + ' '+tmptitle)
        continue
    t3 = json.loads(r3.content.decode())
    t3.update({"artistsrc":tmpartist,'titlesrc':tmptitle})
    #print(t3)
    dfadd = dfadd.append( t3 , ignore_index=True)
    
    # protection erreur 429
    cptr +=1
    if cptr == 10:
        time.sleep(1)
        cptr =0
```
